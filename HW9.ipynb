```python
##question 2:

from random import random
from numpy import arange
from pylab import plot,xlabel,ylabel,show

# Constants
NBi213 = 10000                       # Initial number of bismuth atoms
NTl209 = 0                           # Number of thallium atoms
NPb209 = 0                           # Number of lead atoms
NBi209 = 0                           # Number of bismuth 209 atoms
tau_Bi213 = 46 * 60                  # Half life of bismuth in seconds
tau_Tl209 = 2.2 * 60                 # Half life of thallium in seconds
tau_Pb209 = 3.3 * 60                 # Half life of lead in seconds
h = 1                                # Size of time-step in seconds
p_Bi_decay = 1 - 2**(-h/tau_Bi213)   # Probability of decay in one step
p_Tl_decay = 1 - 2**(-h/tau_Tl209)
p_Pb_decay = 1 - 2**(-h/tau_Pb209)
tmax = 20000                         # Total time

# Lists of plot points
tpoints = arange(0,tmax,h)           # make time array
Bi213points = []                     # empty array to store # of Bi213 atoms at each time step
Tl209points = []                     # empty array to store # of Tl209 atoms at each time step
Pb209points = []                     # empty array to store # of Pb209 atoms at each time step
Bi209points = []                     # empty array to store # of Bi209 atoms at each time step

# Main loop
for t in tpoints:                    # for time 0-1000 sec
    Bi213points.append(NBi213)       # append starting # of Bi213 atoms
    Tl209points.append(NTl209)       # append starting # of Tl209 atoms
    Pb209points.append(NPb209)       # append starting # of Pb209 atoms
    Bi209points.append(NBi209)       # append starting # of Bi209 atoms
    
    
    # Calculate the number of atoms that decay
    decayPb209 = 0
    for a in range(NPb209):          # determine number of atoms that decay
        if random() < p_Pb_decay:
            decayPb209 += 1
    NPb209 -= decayPb209             # subtract number of decayed Parent atoms 
    NBi209 += decayPb209             # add number of decayed Daughter atoms 
    
    decayTl209 = 0
    for b in range(NTl209):
        if random() < p_Tl_decay:
            decayTl209 += 1
    NTl209 -= decayTl209
    NPb209 += decayTl209
    
    decayBi213 = 0

    for c in range(NBi213):
        if random() < p_Bi_decay: 
            #decayBi213 += 1
            NBi213 -= 1
            if random() < 0.0209:
                NTl209 += 1
            else:
                NPb209 += 1
```
